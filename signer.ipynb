```sage
#IMPLEMENTAZIONE SEED TREE ALESSIO
M = 8

#Compute the hash of an input node and separates the result in two parts
def hash_forking(input):

    dig = hash(input) 
    sx = dig[:len(dig)//2]
    dx = dig[len(dig)//2:]

    return sx, dx

##################################################################

#Generates the M seeds as leaves of a binary tree obtained by a starting seed
def SeedTree(starting_seed):

    last_level = [str(starting_seed)]

    while len(last_level) < M:

        next_level = []

        for i in last_level:
            sx, dx = hash_forking(i)
            next_level.append(sx) # Sibling node (SX)
            next_level.append(dx) # Sibling node (DX)
        
        last_level = next_level

    return last_level[0:M] # Return the M required seeds

set_random_seed()
starting_seed = initial_seed()         # Random starting seed, root of the seed tree
seed = SeedTree(starting_seed)

print(starting_seed)
print(seed)
for i in range(0,7):
    print(len(seed[i])*4)
```
